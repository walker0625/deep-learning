## 1. 경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100)
X
```

```python
y = X * 0.7
y
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.show()
```

```python
w = np.random.uniform(-1, 1) # -1 ~ 1까지의 값 랜덤 생성
w
```

```python
b = np.random.uniform(-1, 1)
b
```

```python
lr = 0.5 # learning rate

for epoch in range(200):
    
    y_pred = w * X + b # 식(선형) 자체가 model - 딥러닝(비선형)
    
    error = np.abs(y_pred - y).mean() # 300번의 오차 평균 => 손실함수
    
    # 미분(MAE)
    w_grad = lr * ((y_pred - y) * X).mean() # 현재 가중치에서 수정되어야 할 값
    b_grad = lr * (y_pred - y).mean()
    
    w = w - w_grad # 역전파
    b = b - b_grad
    
    if epoch % 10 == 0:
        print('error : ', error)
        y_pred = w * X + b
        
        plt.figure(figsize=(8, 6))
        plt.scatter(X, y) # 실제(blue)
        plt.scatter(X, y_pred) # 예측(orange)
        plt.show()
        
    if error < 0.001: # early stopping 
        break    
```
